```python
import pandas as pd

# Load the datasets
humidity_df = pd.read_csv(r"C:\Users\mdnih\Downloads\DIA_proj\humidity.csv",encoding='utf-8')
pressure_df = pd.read_csv(r"C:\Users\mdnih\Downloads\DIA_proj\pressure.csv",encoding='utf-8')
temperature_df = pd.read_csv(r"C:\Users\mdnih\Downloads\DIA_proj\temperature.csv",encoding='utf-8')

# Display the first few rows and summary statistics of each dataset

```

```python
# Convert 'datetime' to datetime type and set as index
humidity_df['datetime'] = pd.to_datetime(humidity_df['datetime'])
pressure_df['datetime'] = pd.to_datetime(pressure_df['datetime'])
temperature_df['datetime'] = pd.to_datetime(temperature_df['datetime'])

humidity_df.set_index('datetime', inplace=True)
pressure_df.set_index('datetime', inplace=True)
temperature_df.set_index('datetime', inplace=True)

# Calculate percentage of missing values in each dataset
missing_percentages = {
    'humidity': humidity_df.isnull().mean() * 100,
    'pressure': pressure_df.isnull().mean() * 100,
    'temperature': temperature_df.isnull().mean() * 100
}

missing_percentages
```

```python
# Interpolate missing values for all three datasets
humidity_df = humidity_df.interpolate(method='linear', limit_direction='both')
pressure_df = pressure_df.interpolate(method='linear', limit_direction='both')
temperature_df = temperature_df.interpolate(method='linear', limit_direction='both')

# Check if there are any remaining missing values after interpolation
remaining_missing = {
    'humidity': humidity_df.isnull().sum().sum(),
    'pressure': pressure_df.isnull().sum().sum(),
    'temperature': temperature_df.isnull().sum().sum()
}

remaining_missing
```

```python
import pandas as pd

# Load the data (assuming you already have the CSV files loaded if not, you might need to load them here)
# humidity_df = pd.read_csv('/path/to/humidity.csv')
# pressure_df = pd.read_csv('/path/to/pressure.csv')
# temperature_df = pd.read_csv('/path/to/temperature.csv')

# Print the columns and index name to understand the current dataframe structure
print("Humidity Columns and Index:")
print(humidity_df.columns)
print("Index:", humidity_df.index.name)

print("\nPressure Columns and Index:")
print(pressure_df.columns)
print("Index:", pressure_df.index.name)

print("\nTemperature Columns and Index:")
print(temperature_df.columns)
print("Index:", temperature_df.index.name)

# Check if 'datetime' needs to be set as index or already is the index
if 'datetime' in humidity_df.columns:
    humidity_df.set_index('datetime', inplace=True)
if 'datetime' in pressure_df.columns:
    pressure_df.set_index('datetime', inplace=True)
if 'datetime' in temperature_df.columns:
    temperature_df.set_index('datetime', inplace=True)

# If 'datetime' is not a column, it means it's already the index or there's an issue with the data loading

# Ensure all indices are datetime types
humidity_df.index = pd.to_datetime(humidity_df.index)
pressure_df.index = pd.to_datetime(pressure_df.index)
temperature_df.index = pd.to_datetime(temperature_df.index)

# Rename columns to avoid conflicts
humidity_df.columns = [f"{col}_humidity" for col in humidity_df.columns]
pressure_df.columns = [f"{col}_pressure" for col in pressure_df.columns]
temperature_df.columns = [f"{col}_temperature" for col in temperature_df.columns]

# Merge the dataframes on the datetime index
merged_df = humidity_df.join([pressure_df, temperature_df])

# Display the first few rows of the merged dataframe to check the merge
print("\nMerged DataFrame:")
merged_df.head()
```

```python
# Save the merged dataframe to a CSV file
merged_df.to_csv(r'C:\Users\mdnih\Downloads\DIA_proj\merged_data.csv', index=True)  # Set index=True to include the datetime index in the CSV

print("Merged dataframe saved as 'merged_data.csv'.")
```
