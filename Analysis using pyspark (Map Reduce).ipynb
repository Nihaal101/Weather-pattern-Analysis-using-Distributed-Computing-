```
import sys
from pyspark.sql.window import Window
from pyspark.sql.functions import last, first, col
windowSpec = Window.orderBy("datetime").rowsBetween(-sys.maxsize, 0)
windowSpecReverse = Window.orderBy(col("datetime").desc()).rowsBetween(0, sys.maxsize)
def fill_missing(df):
    columns = df.columns[1:] 
    for column in columns:
        df = df.withColumn(column, last(column, ignorenulls=True).over(windowSpec))
        df = df.withColumn(column, first(column, ignorenulls=True).over(windowSpecReverse))
    return df

humidity_df = fill_missing(humidity_df)
pressure_df = fill_missing(pressure_df)
temperature_df = fill_missing(temperature_df)
```

```
def extract_relevant_temperature(row):
    cities = ['New York', 'Toronto', 'Jerusalem']
    temp_data = []
    for city in cities:
        temp_field = f'{city}_temperature'
        if hasattr(row, temp_field):
            temperature = getattr(row, temp_field)
            if temperature is not None:
                temp_data.append((city, temperature))
    return temp_data


city_temp_pairs = weather_rdd.flatMap(extract_relevant_temperature)
city_temp_mapped = city_temp_pairs.map(lambda x: (x[0], (x[1], 1)))

#  sum of the temperatures and count the occurrences
city_temp_reduced = city_temp_mapped.reduceByKey(lambda a, b: (a[0] + b[0], a[1] + b[1]))

#average temperature for each city
city_average_temp = city_temp_reduced.map(lambda x: (x[0], x[1][0] / x[1][1]))
results = city_average_temp.collect()
for result in results:
    print(f"City: {result[0]}, Average Temperature: {result[1]:.2f}")
```

```
import matplotlib.pyplot as plt
import seaborn as sns

def correlation_analysis(city, df):
    city_specific_columns = [col for col in df.columns if city in col]
    city_data = df.select(city_specific_columns)
    city_data_pd = city_data.toPandas()
    correlation_matrix = city_data_pd.corr()
    
    plt.figure(figsize=(10, 8))
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
    plt.title(f'Correlation Matrix for {city}')
    plt.show()

correlation_analysis('Toronto', weather_data_df)
correlation_analysis('New York', weather_data_df)
correlation_analysis('Jerusalem', weather_data_df)
```

```
def extract_city_data(row):
    cities = ['New York', 'Toronto', 'Jerusalem']
    data = []
    for city in cities:
        temp_key = f'{city}_temperature'
        humidity_key = f'{city}_humidity'
        if hasattr(row, temp_key) and hasattr(row, humidity_key):
            temperature = getattr(row, temp_key)
            humidity = getattr(row, humidity_key)
            if temperature is not None and humidity is not None:
                data.append((city, (float(temperature), float(humidity))))
    return data

city_data_pairs = weather_rdd.flatMap(extract_city_data)

city_data_mapped = city_data_pairs.map(lambda x: (x[0], (x[1][0], x[1][1], x[1][0]*x[1][1], x[1][0]**2, x[1][1]**2, 1)))

city_data_reduced = city_data_mapped.reduceByKey(lambda a, b: (a[0] + b[0], a[1] + b[1], a[2] + b[2], a[3] + b[3], a[4] + b[4], a[5] + b[5]))

def compute_correlation(data):
    from math import sqrt 
    n = data[5]
    sum_xy = data[2]
    sum_x = data[0]
    sum_y = data[1]
    sum_x2 = data[3]
    sum_y2 = data[4]
    numerator = n * sum_xy - sum_x * sum_y
    denominator = sqrt((n * sum_x2 - sum_x**2) * (n * sum_y2 - sum_y**2))
    if denominator == 0:
        return 0
    return numerator / denominator

city_correlation = city_data_reduced.mapValues(compute_correlation)

results = city_correlation.collect()

for result in results:
    print(f"City: {result[0]}, Correlation between Temperature and Humidity: {result[1]:.3f}")
```

```
def extract_pressure_data(row):
    cities = ['New York', 'Toronto', 'Jerusalem']
    data = []
    for city in cities:
        pressure_key = f'{city}_pressure'
        if hasattr(row, pressure_key) and getattr(row, pressure_key) is not None:
            pressure = getattr(row, pressure_key)
            datetime = row.datetime  # Assuming 'datetime' is a column in your DataFrame
            data.append((city, (datetime, float(pressure))))
    return data

city_pressure_data = weather_rdd.flatMap(extract_pressure_data)

def moving_average(numbers, window_size=3):
    i = 0
    moving_averages = []
    while i < len(numbers) - window_size + 1:
        this_window = numbers[i : i + window_size]
        window_average = sum(this_window) / window_size
        moving_averages.append(window_average)
        i += 1
    return moving_averages

# moving averages for pressure
city_grouped = city_pressure_data.groupByKey().mapValues(list)
city_moving_averages = city_grouped.mapValues(lambda x: 
    (list(map(lambda y: y[0], x)), moving_average(list(map(lambda y: y[1], x)))))
city_limited_averages = city_moving_averages.mapValues(lambda x: (x[0][:10], x[1][:10]))


results = city_limited_averages.collect()
for result in results:
    city = result[0]
    timestamps, averages = result[1]
    print(f"City: {city}")
    for time, avg in zip(timestamps, averages):
        print(f"Date: {time}, Moving Average of Pressure: {avg:.2f}")
```
